Locate the Lane Lines
===

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Load our image
# `mpimg.imread` will load .jpg as 0-255, so normalize back to 0-1
img = mpimg.imread('warped_example.jpg')/255

def hist(img):
    # Grab only the bottom half of the image
    # Lane lines are likely to be mostly vertical nearest to the car
    print(img.shape)
    bottom_half = img[img.shape[0]//2:, :]

    # Sum across image pixels vertically - make sure to set `axis`
    # i.e. the highest areas of vertical lines should be larger values
    histogram = np.sum(bottom_half, axis=0)
    
    return histogram

# Create histogram of image binary activations
histogram = hist(img)


# Plot the results
f, (p1, p2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
p1.imshow(img)
p1.set_title('Original Image', fontsize=50)

# Visualize the resulting histogram
p2.plot(histogram)
p2.set_title('Histogram', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)

# Visualize both
# Plot the results
f, (p) = plt.subplots(1, 1, figsize=(24, 9))
p.imshow(img)
p.plot(histogram)
p.set_title('Image with histogram', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

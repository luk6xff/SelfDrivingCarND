```python
%matplotlib inline
import plotly.offline as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import numpy as np
py.init_notebook_mode()
from ipywidgets import interact
```

```python
mean = 0
std = 3
gauss = np.random.normal(mean, std, 10000)
```

```python
def f(x):
    return np.arctan(x)

f = np.vectorize(f)
```

```python
t = np.linspace(-10, 10, 200)
applied = f(gauss)
f_eval = f(t)
```

```python

def f_taylor(x):
    f_0 = np.arctan(0)
    df = 1
    return f_0+df*x

f_taylor = np.vectorize(f_taylor)
f_eval_taylor = f_taylor(t)
applied_linearized = f_taylor(gauss)
```

```python
traces = []
traces.append(go.Histogram(x=gauss,
                           xbins=dict(
                                        start=-30, 
                                        size=0.3, 
                                        end = 30), 
                           histnorm='probability density',
                           name = "gauss",
                           marker=dict(color='rgb(250,150,30)')))


layout = go.Layout(
        xaxis=dict(
        range=[-10, 10],
        dtick=1.0
    )
)
fig = go.Figure(data=traces, layout=layout)
py.iplot(fig)
```

```python
traces = []
traces.append(go.Scatter(y=f_eval, x=t, mode='line'))
traces.append(go.Scatter(y=f_eval_taylor, x=t, mode='line'))
layout = go.Layout(
        xaxis=dict(
        range=[-10, 10],
        dtick=1.0
    ),
    yaxis=dict(
        range=[-1.5, 1.5],
        dtick=1.0
    )
)
fig = go.Figure(data=traces, layout=layout)
py.iplot(fig)
```

```python
traces = []
traces.append(go.Histogram(x=applied,
                           xbins=dict(start=-30, size=0.05, end = 30), 
                           histnorm='probability', 
                           name = "f(gauss)"))
layout = go.Layout(
    xaxis=dict(
        range=[-10, 10],
        dtick=1.0,
    )
)
fig = go.Figure(data=traces, layout=layout)
py.iplot(fig)
```

```python
traces = []
traces.append(go.Histogram(x=applied_linearized,
                           xbins=dict(start=-30, size=0.3, end = 30), 
                           histnorm='probability', 
                           name = "f_taylor(gauss)"))
layout = go.Layout(
    xaxis=dict(
        range=[-10, 10],
        dtick=1.0,
    )
)
fig = go.Figure(data=traces, layout=layout)
py.iplot(fig)
```

```python
def react(t):
    traces = []
    traces.append(go.Scatter(y=f_eval, x=t, mode='line'))
    layout = go.Layout(
            xaxis=dict(
            range=[-10, 10],
            dtick=1.0
        )
    )
    fig = go.Figure(data=traces, layout=layout)
    py.iplot(fig)
interact(react, t=(0.0,5.0), add=(-10.0, 10.0));
```

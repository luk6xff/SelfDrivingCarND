```python
%matplotlib inline
import plotly.offline as py
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import numpy as np
py.init_notebook_mode()
from ipywidgets import interact
```

```python
mean = 0
std = 3
gauss = np.random.normal(mean, std, 10000)
```

```python
def f(x):
    return np.arctan(x)

f = np.vectorize(f)
applied = f(gauss)
    
```

```python
def react(factor,add):
    traces = []
    applied = factor*gauss + add
    traces.append(go.Histogram(x=applied, histnorm='density', name = "f(gauss)"))
    traces.append(go.Histogram(x=gauss, histnorm='density',name = "gauss"))
    py.iplot(traces)
```

```python
interact(react, factor=(0.0,5.0), add=(-10.0, 10.0));
```
